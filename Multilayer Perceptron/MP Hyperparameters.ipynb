## Activation Function= Type of non-linearity introduced to the model


The first on the left here is a sigmoid curve, otherwise known as a logistic curve. And that's actually what it's referred to in scikit-learn. But we see that familiar S shaped curve from zero to one that looks exactly like what we saw with logistic regression.

And then the middle here, there's the hyperbolic tangent curve, which is referred to as tan H. This is also an S shape, but it's between negative one and one. 

Lastly, and probably the most popular is called RELU. RELU stands for rectified linear unit, and it's actually very simple. It just sets a floor at zero, so if any numbers below zero, it sets it to zero. And if it's above zero, then it doesn't do anything. For such a simple function, it's surprisingly powerful.

## Learning_rate= how quickly and whether or not the algorithm will find the optimal solution


A high learning rate will push the algorithm along quicker to find the optimal solution, but it's possible that it pushes it too far and it never actually finds the best solution. On the other end, a lower learning rate will take much longer for the algorithm to fit, but you can have more confidence that it will find the optimal solution. 

This curve represents the loss curve with the bottom representing the optimal model with the lowest error rate. On the left, the model is taking steps that are too large in order to find the best solution. And it actually will never find the optimal solution. And on the right, it will find the optimal solution, but it will take a while because it's taking such small steps each time. So the goal is to find the right balance between where it will find the optimal solution, but it won't take too long to do so.
